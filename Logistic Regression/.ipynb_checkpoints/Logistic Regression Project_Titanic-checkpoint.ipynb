## Part 2

Time to predict the missing ages using class

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='Pclass',y='Age',data=train)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age

# cols[0] first item, cols[1] second item    
# if there is no age assigned look at each class and fill in using the average
```

```python
train['Age'] = train[['Age','Pclass']].apply(impute_age,axis=1)

# axis = 1 is columns
```

In the function Age = cols[0] links to the 'Age' part within the list when training. While, Pclass = cols[1] lins to the 'Pclass' part within the list during the training.

It will pull the 'Age' and 'Pclass' from the data base then the [0] and [1] will be input into the function to keep it in order.

Then goes to pd.isnull(Age), if the Pclass is 1,2,3 it will fill in the age if there is not one given already.

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')

# age is all filled out!
# can always train a function to predict age too
```

```python
train.drop('Cabin',axis=1,inplace=True)

# so much missing cabin information it's not worth it to fill out
```

```python
train.dropna(inplace=True)

# drops all of the other missing data
```

**now that the data is clean we will need to convert the categorical features into a dummy variable using pandas.** If it's not done the machine learning algorithm won't be able to directly take in those features as inputs. 

ex. 'Sex' = 'male' or 'female'. Needs to become 0 or 1

```python
pd.get_dummies(train['Sex'])

# converts categorical variables into dummy or indicator variables
# gets data frame, 0 if false 1 if true

# the algorithm if fed both columns will predict perfectly that if one is 0-
# -the other must be 1. This is called multicolinearity and messes up the 
# algorithm
```

```python
# to avoid this we must drop one column

sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex.head(2)
```

```python
embark = pd.get_dummies(train['Embarked'])
```

```python
embark.head()
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark.head()

# dropping the C column made it so they're no longer perfect predictors
```

```python
train = pd.concat([train,sex,embark],axis=1)
```

```python
train.head()

# added 'male','Q','S' columns...our replacement columns!
# no longer 'embark', 'sex'
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis = 1, inplace=True)
```

```python
train.head()

# all numerical
```

**the absensce of Q or S means C, so it is not needed in the equation**

**if more than 2 categories always drop one to avoid Multicollinearity**

```python
train.drop('PassengerId',axis=1,inplace=True)
```

## Part 3


time to test and train the data

```python
# y is the column we are trying to predict

X = train.drop('Survived', axis = 1)
y = train['Survived']
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

Use a classification report 

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```
